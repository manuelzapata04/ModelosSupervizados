```python
from functions_ok import *
import pickle

pkl_filename = "pickle_nasa.pkl"
with open(pkl_filename, 'rb') as file:
    pickle_nasa = pickle.load(file)
```

# Lectura de Datos

```python
#Leer datos
#cambiar archivo para datos no vistos
df=  pd.read_csv("nasa.csv").sample(100)
```

# Prediciones

```python
# Calcula las predicciones
pred = pickle_nasa.predict(df)
pred
```

```python
#calcula las probas
probas=pickle_nasa.predict_proba(df)[:,1]
probas
```

```python
#añade las prediciones y sus probas al df
df["prediccion"] = pred
df["probabilidad"] = probas
df
```
